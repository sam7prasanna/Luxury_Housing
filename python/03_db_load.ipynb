```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/luxury_housing_features.csv")
df.shape
```

```python
username = "root"
password = "root"
host = "localhost"
port = "3306"
database = "luxury_housing_db"

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
)
```

```python
df.to_sql(
    name="luxury_housing_sales",
    con=engine,
    if_exists="append",
    index=False,
    chunksize=1000
)
print("Data load completed successfully")
```
